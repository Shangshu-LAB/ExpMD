```python
import pandas as pd
import numpy as np
import json
import gzip
```

```python
import torch
```

```python
dataset_name = '500'
```

```python
with gzip.open(f'Dataset/dataset_{dataset_name}.json.gz') as f:
    DATA = json.loads(f.read())
df = pd.DataFrame(DATA)
df
```

```python
with open(f'Dataset/label_2100.info', encoding='utf-8') as f:
    label2idx = eval(f.read())
labels = list(label2idx.keys())
label2idx
```

```python
df_known = df[df['label'].isin(label2idx.keys())]
df_unknown = df[~df['label'].isin(label2idx.keys())]
df_known.shape, df_unknown.shape
```

```python
labels_ = list(df_unknown['label'].value_counts().index)
label2idx_ = dict(zip(labels_,range(len(labels_))))
len(label2idx_)
```

```python
y_test = df_known['label'].apply(lambda x:label2idx[x]).to_numpy()
```

```python
# rootdir_cnn = 'model_trained/clsText-ResNet_202307140956'
# rootdir_rnn = 'model_trained/multiPS-RNN_202307230353'

rootdir_cnn = 'model_trained/Text-Net'
rootdir_rnn = 'model_trained/PS-Net'
```

```python
from Text_Net import ResNet
from PS_Net import RNN
import os
import joblib

device = torch.device('cuda' if torch.cuda.is_available() else 'cpu')

model_text = ResNet(class_num=104,out_dim=16,block_num=2).to(device)
model_text.load_state_dict(torch.load(os.path.join(rootdir_cnn,'best.pt'),map_location=device))
model_text.eval()

model_ps = RNN(class_num=104,out_dim=16,hidden_size=128).to(device)
model_ps.load_state_dict(torch.load(os.path.join(rootdir_rnn,'best.pt'),map_location=device))
model_ps.eval()
```

```python
def padding(load, MAXLEN=784, padding_idx=0x100):
    if len(load)>=MAXLEN:
        return load[0:MAXLEN]
    else:
        return load+[padding_idx]*(MAXLEN-len(load))
```

```python
from tqdm import tqdm
```

```python
X_upload = df_known['upload'].apply(padding).tolist()
X_download = df_known['download'].apply(padding).tolist()

p_text = []
BATCH_SIZE = 128
for i in tqdm(range(0,len(df_known),BATCH_SIZE)):
    tmp = model_text(
        x_up=torch.tensor(X_upload[i:i+BATCH_SIZE]).long().to(device),
        x_down=torch.tensor(X_download[i:i+BATCH_SIZE]).long().to(device)
    ).detach().tolist()
    p_text += tmp
p_text = np.array(p_text)

p_text.shape
```

```python
from sklearn.metrics import accuracy_score, confusion_matrix, classification_report

y_pred = p_text.argmax(axis=1)

print(accuracy_score(y_true=y_test,y_pred=y_pred))
print(confusion_matrix(y_true=y_test,y_pred=y_pred))
print(classification_report(y_true=y_test,y_pred=y_pred,digits=4,target_names=labels))
```

```python
pd.DataFrame(confusion_matrix(y_true=y_test,y_pred=y_pred),index=labels,columns=labels).to_csv(f'result\\text_CM.csv')
pd.DataFrame(classification_report(y_true=y_test,y_pred=y_pred,target_names=labels,output_dict=True)).transpose().to_csv(f'result\\text_report.csv')
```

```python
from torch.nn.utils.rnn import pad_sequence

MAXLEN =256

path = df_known['packets'].apply(lambda x:torch.tensor(
    [[p['iat'], p['l'], p['b'], 1 if p['d'] == '>' else -1] for i,p in enumerate(x[0:MAXLEN])]
)).tolist()
path_len = df_known['packets'].apply(lambda x:min(len(x),MAXLEN)).tolist()

p_ps = []
for i in tqdm(range(0,len(df_known),BATCH_SIZE)):
    tmp = model_ps(
        path=pad_sequence(path[i:i+BATCH_SIZE],batch_first=True).float().to(device),
        path_len=torch.tensor(path_len[i:i+BATCH_SIZE])
    ).detach().tolist()
    p_ps += tmp
p_ps = np.array(p_ps)
p_ps.shape
```

```python
y_pred = p_ps.argmax(axis=1)

print(accuracy_score(y_true=y_test,y_pred=y_pred))
print(confusion_matrix(y_true=y_test,y_pred=y_pred))
print(classification_report(y_true=y_test,y_pred=y_pred,digits=4,target_names=labels))
```

```python
pd.DataFrame(confusion_matrix(y_true=y_test,y_pred=y_pred),index=labels,columns=labels).to_csv(f'result\\ps_CM.csv')
pd.DataFrame(classification_report(y_true=y_test,y_pred=y_pred,target_names=labels,output_dict=True)).transpose().to_csv(f'result\\ps_report.csv')
```
