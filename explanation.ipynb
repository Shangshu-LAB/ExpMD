```python
import pandas as pd
import numpy as np
import json
import gzip
import os
```

```python
dataset_name = '500'
# dataset_name = 'sample'
```

```python
with gzip.open(f'Dataset/dataset_{dataset_name}.json.gz') as f:
    DATA = json.loads(f.read())
df = pd.DataFrame(DATA)
df
```

```python
with open(f'Dataset/label_2100.info', encoding='utf-8') as f:
    label2idx = eval(f.read())
labels = list(label2idx.keys())
label2idx
```

```python
rootdir_cnn = 'model_trained/Text-Net'
rootdir_rnn = 'model_trained/PS-Net'
rootdir_base = 'model_trained/Meta-RF.joblib'
rootdir_BD = 'model_trained/BD-boost.joblib'
```

```python
from Text_Net import ResNet
from PS_Net import RNN
import os
import joblib
import torch

device = torch.device('cuda' if torch.cuda.is_available() else 'cpu')
model_text = ResNet(class_num=104,out_dim=16,block_num=2)
model_text.load_state_dict(torch.load(os.path.join(rootdir_cnn,'best.pt'),map_location=device))
model_text.eval()

model_ps = RNN(class_num=104,out_dim=16,hidden_size=128)
model_ps.load_state_dict(torch.load(os.path.join(rootdir_rnn,'best.pt'),map_location=device))
model_ps.eval()

model_base = joblib.load(rootdir_base)
model_BD = joblib.load(rootdir_BD)
```

```python
def padding(load, MAXLEN=784, padding_idx=0x100):
    if len(load)>=MAXLEN:
        return load[0:MAXLEN]
    else:
        return load+[padding_idx]*(MAXLEN-len(load))
```

```python
import matplotlib.pyplot as plt
from matplotlib.colors import Normalize
import seaborn as sns
import math
```

```python
samples = pd.DataFrame()
# for label in labels[0:10]:
for label in labels:
    df_tmp = df[(df['label']==label)&(df['dp']!=443)&(df['dp']!=80)]
    if len(df_tmp)>8:
        sample = df_tmp.sample(n=8)
        samples = pd.concat([samples, sample])
    else:
        continue
samples
```

```python
y_true = samples['label'].apply(lambda x:label2idx[x]).tolist()
print(y_true)
```

```python
features = ['bytes_out','pkts_out','bytes_in','pkts_in','dur']
X_base = samples[features]

p_base = model_base.predict_proba(X=X_base)
r_base = p_base.argmax(axis=1)
print(r_base)
```

```python
import shap
# 创建一个 SHAP 解释器
explainer_base = shap.TreeExplainer(model_base)
# 获取单个样本的 SHAP 值
shap_values_base = explainer_base.shap_values(X_base)
```

```python
X_upBD, X_downBD = np.array(samples['upBD'].tolist()), np.array(samples['downBD'].tolist())
X_upBD_ = np.nan_to_num(X_upBD / X_upBD.sum(axis=1)[:, np.newaxis])
X_downBD_ = np.nan_to_num(X_downBD / X_downBD.sum(axis=1)[:, np.newaxis])

X_BD = np.concatenate([X_upBD_, X_downBD_], axis=1)

p_BD = model_BD.predict_proba(X=X_BD)
r_BD = p_BD.argmax(axis=1)
print(r_BD)
```

```python
explainer_BD = shap.TreeExplainer(model_BD)
shap_values_BD = explainer_BD.shap_values(X_BD)
```

```python
upload = torch.tensor(samples['upload'].apply(padding).tolist())
download = torch.tensor(samples['download'].apply(padding).tolist())
upload_len = samples['upload'].apply(lambda x:len(x)).tolist()
download_len = samples['download'].apply(lambda x:len(x)).tolist()

upload_emb = model_text.embedding(upload)
download_emb = model_text.embedding(download)

p_Text = model_text.forward_from_embedding(upload_emb,download_emb)
r_Text = p_Text.max(dim=1)
print(r_Text.indices)
```

```python
from tqdm import tqdm
from captum.attr import IntegratedGradients

explainer_Text = IntegratedGradients(model_text.forward_from_embedding)

# attr_Textup, attr_Textdown = explainer_Text.attribute(
#     inputs=(upload_emb,download_emb),
#     target=y_true
#     # target=r_Text.indices
# )
attr_Textup, attr_Textdown = [], []
BATCH_SIZE = 16
for i in tqdm(range(0,len(samples),BATCH_SIZE)):
    tmp_up ,tmp_down = explainer_Text.attribute(
        inputs=(upload_emb[i:i+BATCH_SIZE],download_emb[i:i+BATCH_SIZE]),
        target=y_true[i:i+BATCH_SIZE]
    )
    attr_Textup += tmp_up.detach().tolist()
    attr_Textdown += tmp_down.detach().tolist()
attr_Textup = np.array(attr_Textup)
attr_Textdown = np.array(attr_Textdown)

attr_Textup.shape, attr_Textdown.shape
```

```python
from torch.nn.utils.rnn import pad_sequence

MAXLEN =256
path = samples['packets'].apply(lambda x:torch.tensor(
    [[p['iat'], p['l'], p['b'], 1 if p['d'] == '>' else -1] for i,p in enumerate(x[0:MAXLEN])]
)).tolist()
path = pad_sequence(path,batch_first=True).float().to(device)
path_emb = model_ps.get_embedding(path)

path_len = samples['packets'].apply(lambda x:min(len(x),MAXLEN)).tolist()
path_len = torch.tensor(path_len)

p_PS = model_ps.forward_from_embedding(
    path_emb=path_emb, path_len=path_len
)
r_PS = p_PS.max(dim=1)
print(r_PS.indices)
```

```python
explainer_PS = IntegratedGradients(model_ps.forward_from_embedding)

attr_PS = []
for i in tqdm(range(0,len(samples),BATCH_SIZE)):
    tmp = explainer_PS.attribute(
        inputs=path_emb[i:i+BATCH_SIZE],
        additional_forward_args=path_len[i:i+BATCH_SIZE],
        target=y_true[i:i+BATCH_SIZE],
    )
    attr_PS += tmp.detach().tolist()
attr_PS = np.array(attr_PS)

attr_PS.shape
```

```python
def attrmap_BD(X_BD, attr_weight, title):
    max_weight = np.abs(attr_weight).max()
    plt.figure(figsize=(11,5))
    plt.subplot(1,2,1)
    sns.heatmap(
        attr_weight[0:256].reshape((0x10,0x10)),
        annot=(X_BD[0:256]*1000).reshape((0x10,0x10)), annot_kws={'size':5}, fmt='.1f',
        cmap='RdBu_r',vmax=max_weight, vmin=-max_weight,
        cbar_kws={'label':'Attribution Weight'},
        square=True, linewidths=0.5,
    )
    plt.axhline(y=2, color='black', linewidth=0.8)
    plt.axhline(y=8, color='black', linewidth=0.8)
    plt.xticks(np.arange(0,0x10)+0.5,[hex(i)[2:].upper() for i in range(0x10)])
    plt.xlabel('Low Byte')
    plt.yticks(np.arange(0,0x10)+0.5,[hex(i)[2:].upper() for i in range(0x10)],rotation=0)
    plt.ylabel('High Byte')
    plt.title("Uplink BD")

    plt.subplot(1,2,2)
    sns.heatmap(
        attr_weight[256:512].reshape((0x10,0x10)),
        annot=(X_BD[256:]*1000).reshape((0x10,0x10)), annot_kws={'size':5}, fmt='.1f',
        cmap='RdBu_r',vmax=max_weight, vmin=-max_weight,
        cbar_kws={'label':'Attribution Weight'},
        square=True, linewidths=0.5
    )
    plt.axhline(y=2, color='black', linewidth=1.)
    plt.axhline(y=8, color='black', linewidth=1.)
    plt.xticks(np.arange(0,0x10)+0.5,[hex(i)[2:].upper() for i in range(0x10)])
    plt.xlabel('Low Byte')
    plt.yticks(np.arange(0,0x10)+0.5,[hex(i)[2:].upper() for i in range(0x10)],rotation=0)
    plt.ylabel('High Byte')
    plt.title("Downlink BD")

    plt.suptitle(title,fontproperties="Microsoft YaHei")
    plt.tight_layout()
```

```python
def get_stream(load):
    streams = []
    for b in load.tolist():
        if 0<=b<=0xFF:
            if len(chr(b))>1:
                streams.append(hex(b))
            else:
                streams.append(chr(b))
        else:
            streams.append('')
    streams = np.array(streams)
    return streams

def attrmap_Text(upload, download, upload_attr_weight, download_attr_weight, upload_length, download_length, title):
    max_weight = max(np.abs(upload_attr_weight).max(),np.abs(download_attr_weight).max())
    width = 28

    plt.figure(figsize=(9, 4))
    plt.subplot(1, 2, 1)
    height = math.ceil(upload_length/width)
    streams = get_stream(upload)
    sns.heatmap(
        upload_attr_weight[0:width*height].reshape(height,width),#.detach(),
        annot=streams[0:width*height].reshape(height,width),
        annot_kws={'size':6}, fmt='',
        cmap='RdBu_r', vmax=max_weight, vmin=-max_weight,
        cbar_kws={'label':'Attribution Weight'},
        square=True,
    )
    plt.xticks(np.arange(0,width)+0.5,np.arange(0,width),fontsize=5)
    plt.yticks(np.arange(0,height)+0.5,np.arange(0,height), fontsize=5, rotation=0)
    plt.title("Uplink Load")

    plt.subplot(1, 2, 2)
    height = math.ceil(download_length/width)
    streams = get_stream(download)
    sns.heatmap(
        download_attr_weight[0:width*height].reshape(height,width),#.detach(),
        annot=streams[0:width*height].reshape(height,width),
        annot_kws={'size':6}, fmt='',
        cmap='RdBu_r', vmax=max_weight, vmin=-max_weight,
        cbar_kws={'label':'Attribution Weight'},
        square=True,
    )
    plt.xticks(np.arange(0,width)+0.5,np.arange(0,width),fontsize=5)
    plt.yticks(np.arange(0,height)+0.5,np.arange(0,height), fontsize=5, rotation=0)
    plt.title("Downlink Load")

    plt.suptitle(title, fontproperties="Microsoft YaHei")
    plt.tight_layout()
```

```python
def attrmap_PS_(pakcets, attr_weight, length, title):
    pakcets = pd.DataFrame(pakcets)
    # delta_t_min, delta_t_max = pakcets['t'].diff().min()+1e-6, pakcets['t'].diff().max()
    delta_t_min = 1e-3
    data = []
    t_adj = 0
    for i in range(length):
        if i > 0:
            # t_adj += max(math.sqrt((pakcets.iloc[i]['t']-pakcets.iloc[i-1]['t'])/delta_t_min),1)
            t_adj += max(math.log((pakcets.iloc[i]['t']-pakcets.iloc[i-1]['t'])/delta_t_min),1)
        item = {
            't': pakcets.iloc[i]['t'],
            't_adj': t_adj,
            'p': pakcets.iloc[i]['l'] if pakcets.iloc[i]['d']=='>' else -pakcets.iloc[i]['l'],
            'b': pakcets.iloc[i]['b'] if pakcets.iloc[i]['d']=='>' else -pakcets.iloc[i]['b'],
            'attr': attr_weight[i].sum()
        }
        data.append(item)
    data = pd.DataFrame(data)

    sns.set_style("white")
    # 创建一个颜色映射（colormap），用于将权重映射到颜色
    cmap = plt.get_cmap('coolwarm')
    # cmap = plt.get_cmap('RdBu_r')

    # 创建一个子图
    # fig, ax = plt.subplots(figsize=(np.sqrt(data['t_adj'].max()+1), 3))
    fig, ax = plt.subplots(figsize=(10, 3))
    # fig, ax = plt.subplots(figsize=(len(data)/3, 5), dpi=300)
    # fig, ax = plt.subplots()

    # 创建归一化对象，以将权重映射到0到1之间
    max_weight = data['attr'].abs().max()
    norm = Normalize(vmin=-max_weight, vmax=max_weight)

    # 绘制柱状图，设置柱形的高度为数值，颜色为归一化后的权重
    bars = ax.bar(data['t_adj'], data['p'], color=cmap(norm(data['attr'])), width=0.9,linewidth=0)
    # bars = ax.bar(range(len(data)), data['p'], color=cmap(norm(data['attr'])), width=0.9,linewidth=0)
    # 添加颜色标签（colorbar），设置区间
    cbar = plt.colorbar(plt.cm.ScalarMappable(cmap=cmap, norm=norm), ax=ax)
    cbar.set_label('Attribution Weight')

    ax.axhline(y=0, color='black', linewidth=1.)

    ax.set_xlabel(f'Time sqrt(Δt/min(Δt)) min(Δt)={delta_t_min:g}')
    ax.set_xlim([-1,data['t_adj'].max()+1])
    # ax.set_xlim([-1,len(data)+1])

    ax.set_yscale('function',functions=(lambda x:np.cbrt(x),lambda x:np.power(x,3)))
    ax.set_yticks([-1500,-1000,-500,-60,0,60,500,1000,1500])
    ax.set_ylim([-1600,1600])
    ax.set_ylabel('packet length')

    plt.tight_layout()
    plt.title(title,fontproperties="Microsoft YaHei")


def attrmap_PS(pakcets, attr_weight, length, title):
    data = []
    pakcets = pd.DataFrame(pakcets)
    for i in range(length):
        item = {
            't': pakcets.iloc[i]['t'],
            'p': pakcets.iloc[i]['l'] if pakcets.iloc[i]['d']=='>' else -pakcets.iloc[i]['l'],
            'attr': attr_weight[i].sum()
        }
        data.append(item)
    data = pd.DataFrame(data)

    # style must be one of white, dark, whitegrid, darkgrid, ticks
    sns.set_style("white")
    cmap = plt.get_cmap('coolwarm')

    # fig, ax = plt.subplots(figsize=(10, 3))
    fig, ax = plt.subplots(figsize=(int(np.sqrt(length)*1.5)+1, 3))
    max_weight = data['attr'].abs().max()
    # max_length = max(data['p'].abs().max(),2000)
    max_length = 1600
    heatmap_width = 6
    norm = Normalize(vmin=-max_weight, vmax=max_weight)
    ax.bar(range(length), heatmap_width, color=cmap(norm(data['attr'])), width=1,linewidth=0,bottom=-heatmap_width/2)
    cbar = plt.colorbar(plt.cm.ScalarMappable(cmap=cmap, norm=norm), ax=ax)
    cbar.set_label('Attribution Weight')
    cbar.outline.set_linewidth(0.)

    start_point = (-0.5, 0)
    end_point = (length-0.5, 0)
    arrow = plt.arrow(
        *start_point, *(end_point[0] - start_point[0], end_point[1] - start_point[1]),
        edgecolor='black', facecolor='none', linewidth=0.5, width=heatmap_width,
        head_length=3,head_width=heatmap_width*50
    )
    ax.add_patch(arrow)

    # plt.plot([-0.5,length-0.5],60,color='gray', linewidth=0.5)
    # plt.plot([-0.5,length-0.5],[60,60],color='gray', linewidth=0.5)
    plt.hlines(y=[-60,60], xmin=-0.5,xmax=length-0.5, color='gray', linewidth=0.5)
    # plt.axhline(y=-60, color='gray', linewidth=0.5)

    ax.bar(range(length), data['p'].apply(lambda x:max(x,0)), color='black', width=0.5,linewidth=0,bottom=heatmap_width/2)
    ax.bar(range(length), data['p'].apply(lambda x:min(x,0)), color='black', width=0.5,linewidth=0,bottom=-heatmap_width/2)

    def y_function(x):
        # return np.where(x>0,np.sqrt(x),-np.sqrt(-x))
        return np.piecewise(
            x,[x>0,x<0],
            [lambda x:np.cbrt(x)+heatmap_width/2, lambda x:np.cbrt(x)-heatmap_width/2,0]
        )
    def y_function_inverse(x):
        # return np.where(x>0,np.square(x),-np.square(x))
        return np.piecewise(
            x, [x>0,x<0],
            [lambda x:np.power(x-heatmap_width/2,3),lambda x:-np.power(x+heatmap_width/2,3),0]
        )

    ax.set_yscale('function',functions=(y_function,y_function_inverse))
    yticks = np.array([-1500,-1000,-500,-60,-0.0000001,0.0000001,60,500,1000,1500])
    ax.set_yticks(yticks)
    ax.set_ylim([-int(max_length)-1,int(max_length)+1])
    ax.set_ylabel('Packet Length')

    ax.set_xlim([-0.5,length+2.6])
    ax.set_xlabel('Packet Sequence')

    ax.spines['top'].set_visible(False)
    ax.spines['bottom'].set_visible(False)
    ax.spines['right'].set_visible(False)

    plt.tight_layout()
    plt.title(title,fontproperties="Microsoft YaHei")
    # plt.show()
    # plt.close()
```
